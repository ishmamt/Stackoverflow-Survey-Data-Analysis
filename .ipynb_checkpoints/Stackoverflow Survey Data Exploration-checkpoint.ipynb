```python
c_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()).sort_values(ascending=False)
```

```python
c_grp['LanguageWorkedWith'].apply(lambda x: (x.str.contains('Python').sum()*100) / len(x)).head(10)  
# count not used here as it ignores nan
```

```python
# another way to do this using concat
total = c_grp['Country'].count()
py = c_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
py_df = pd.concat([total, py], axis='columns')
py_df.rename(columns = {'LanguageWorkedWith': 'WorkedWithPython', 'Country':'Total'}, inplace=True)
py_df['%Python'] = (py_df['WorkedWithPython'] / py_df['Total']) * 100
py_df.head(10)
```

```python
filt = (py_df['Total'] > 500)
py_df[filt].sort_values(by='%Python', ascending=False).head(10)
```
